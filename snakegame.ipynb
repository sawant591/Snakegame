```python
#import libraries
import turtle
import random
import time

delay=0.1
score=0
high_score=0

#we defined our window
winpanel=turtle.Screen()
winpanel.title("Snake game by Sawant Raj")
winpanel.bgcolor("blue")
winpanel.setup(width=600,height=600)
winpanel.tracer(0)
winpanel.bgpic("D:\\pythongame\\back.png")


#we defined the head of our snake
head=turtle.Turtle()
head.shape("square")
head.color("white")
head.penup()
head.goto(0,0)
head.direction="stop"

#we defined the food
fooditem=turtle.Turtle()
colors=random.choice(['red','green','blue'])
shapes=random.choice(['square','circle','triangle'])
fooditem.shape(shapes)
fooditem.color(colors)
fooditem.penup()
fooditem.speed(0)
fooditem.goto(0,100)

pen=turtle.Turtle()
pen.speed(0)
pen.shape("square")
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,250)
pen.write("Score : 0 High Score : 0" , align="center" , font=("arial",20,"bold"))

#function of the snake

def goup():
    if head.direction !="down":
        head.direction="up"

def godowinpanel():
    if head.direction !="up":
        head.direction="down"

def goleft():
    if head.direction !="right":
        head.direction="left"

def goright():
    if head.direction !="left":
        head.direction="right"
         
def move():
    if head.direction=="up":
        y=head.ycor()
        head.sety(y+20)
    if head.direction=="down":
        y=head.ycor()
        head.sety(y-20)
    if head.direction=="left":
        x=head.xcor()
        head.setx(x-20)
    if head.direction=="right":
        x=head.xcor()
        head.setx(x+20)
        
winpanel.listen()
winpanel.onkeypress(goup,"Up")
winpanel.onkeypress(godowinpanel,"Down")
winpanel.onkeypress(goleft,"Left")
winpanel.onkeypress(goright,"Right")

segments=[]

while True:
    winpanel.update()
    if head.xcor()>290 or head.xcor()<-290 or head.ycor()>290 or head.ycor()<-290:
        time.sleep(1)
        head.goto(0,0)
        head.direction="stop"
        colors=random.choice(['red','green','blue'])
        shapes=random.choice(['square','circle','triangle'])
        fooditem.shape(shapes)
        fooditem.color(colors)
        
        for segment in segments:
            segment.goto(1000,1000)
            segments.clear()
            score=0
            delay=0.1
            pen.clear()
            pen.write("Score : {} High Score : {}".format(score,high_score) , align="center" , font=("arial",20,"bold"))

    #head is attacking the food or not              
    if head.distance(fooditem)<20:
        x=random.randint(-270,270)
        y=random.randint(-270,270) 
        colors=random.choice(['red','green','blue'])
        shapes=random.choice(['square','circle','triangle'])
        fooditem.shape(shapes)
        fooditem.color(colors)
        fooditem.goto(x,y)
     
    #increase the snake size=increase the number of segments
        new_segment=turtle.Turtle()
        new_segment.speed(0)
        new_segment.color("orange")
        new_segment.shape("square")
        new_segment.penup()
        segments.append(new_segment)
        delay-=0.001
        score+=10
        if score>high_score:
            high_score=score
        pen.clear()
        pen.write("Score : {} High Score : {}".format(score,high_score) , align="center" , font=("arial",20,"bold"))

    #checking for head collission with the body segments
    for index in range(len(segments)-1,0,-1):
        x=segments[index-1].xcor()
        y=segments[index-1].ycor()
        segments[index].goto(x,y)
    if len(segments)>0:
        x=head.xcor()
        y=head.ycor()
        segments[0].goto(x,y)
    move()
    for segment in segments:
        if segment.distance(head)<20:
            time.sleep(1)
            head.goto(0,0)
            head.direction="stop"
            for segment in segments:
                segment.goto(1000,1000)
            segment.clear()

            score=0
            delay=0.1
            pen.clear()
            pen.write("Score : {} High Score : {}".format(score,high_score) , align="center" , font=("arial",20,"bold"))
    time.sleep(delay)

winpanel.mainloop()
```
